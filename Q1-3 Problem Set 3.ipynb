# Question 1

### Step 1

```python
import numpy as np, pandas as pd
```

### Step 2, 3
Importing dataset  
Assigning to the variable 'users'

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
users.head()
```

### Step 4
Mean Age per occupation

```python
mean_age_per_occupation = users.groupby(['occupation']).age.mean()
mean_age_per_occupation
```

### Step 5
Male ratio per occupation from the most to least

```python
users['males'] = users.gender.apply(lambda m: True if m=='M' else False)
```

```python
male_count = users.groupby('occupation')['males'].sum()
total_gender_count = users.groupby('occupation')['gender'].count()
```

```python
male_ratio = male_count / total_gender_count
male_ratio.sort_values(ascending = False)
```

### Step 6
Min and Max age for each occupation

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7
Mean Age for each combination of occupation and sex

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### Step 8
Percentage of women and men for each occupation

```python
countOf_malesANDfemales = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occupation_count = users.groupby(['occupation']).count()
gender_percentage = countOf_malesANDfemales.div(occupation_count, level = "occupation")
round(gender_percentage['gender']*100,2)
```
